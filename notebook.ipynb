### - BFS Report

BFS Explores all nodes at the present depth level before moving deeper

Advantages:
- **Complete**: Always finds solution if exists
- **Optimal**: Finds shortest path (minimum number of moves)
- **Systematic**: Explores all possibilities at each depth


Disadvantages:
- **Memory intensive**: O(b^d) space complexity
- Stores all nodes at current depth level
- Can be slow for deep solutions


Data Structures Used:
- **Queue (deque)**: Frontier of nodes to explore (FIFO)
- **Set**: Explored states to prevent revisiting
- **Node Class**: Stores state, parent, move, depth, cost

## 2- DFS

```python
dfs(initial_state)
```

### - DFS Report

DFS Explores as deep as possible along each branch before backtracking


Advantages:
- **Memory efficient**: O(bm) where m is max depth
- Space complexity much better than BFS
- Can find solution quickly if it's deep in the tree


Disadvantages:
- **Not optimal**: May not find shortest path
- **Not complete**: Can get stuck in infinite paths
- May explore very deep before finding solution
- Can go down wrong path for a long time


Data Structures Used:
- **Stack**: Frontier of nodes to explore (LIFO)
- **Set**: Explored states to avoid cycles
- **Node Class**: Stores state, parent, move, depth, cost


Depth-Limited DFS:
- Adds depth limit parameter to prevent infinite exploration
- Truncates search at specified depth
- Useful as building block for IDFS

## 3- Iterative DFS

```python
idfs(initial_state)
```

### - IDFS Report

IDFS Combines DFS's memory efficiency with BFS's completeness
- Runs DFS repeatedly with increasing depth limits
- Stops when solution found or max depth reached
- Each iteration starts from scratch with deeper limit


Advantages:
- **Complete**: Always finds solution if exists
- **Optimal**: Finds shortest path for uniform cost
- **Memory efficient**: O(bd) space (like DFS)
- **Better than BFS**: For deep solutions, uses less memory
- **Better than DFS**: Guarantees shortest path


Disadvantages:
- **Repeated work**: Regenerates nodes at each iteration
  - Root node regenerated d times
  - Level 1 nodes regenerated d-1 times, etc.
- Can be slower than BFS for shallow solutions
- Need to choose appropriate max depth


Data Structures Used:
- **Stack**: For depth-limited DFS at each iteration
- **Set**: Explored states for current iteration
- **Counter**: Track total nodes expanded across iterations


**Why IDFS is better than DFS:**
- Guarantees finding shortest path
- Avoids infinite paths
- Still memory efficient

**Why IDFS is better than BFS for deep solutions:**
- Same time complexity as BFS (O(b^d))
- Much better space complexity: O(bd) vs O(b^d)

###  - A* Report
A* finds the shortest path to solve the puzzle by combining:

- g(n) = moves made so far

- h(n) = estimated moves remaining (heuristic)

- f(n) = g(n) + h(n) = total estimated cost

Advantages of A*:

- Optimal: Finds shortest path (with admissible heuristic)

- Complete: Always finds solution if exists

- Efficient: Uses heuristic to guide search toward goal

#### Data Structures Used

**Priority Queue (Heap)** : Frontier of nodes to explore -> always expands lowest f-cost

**Set** : Explored states -> prevents revisiting states

**Node Class** : Stores state, parent, move, depth, cost -> easy path reconstruction and tracking

**Counter** : Tie-breaking in heap -> prevents comparing Node objects directly

#### Heuristic Functions
Manhattan Distance:

- Admissible: Never overestimates actual cost

- Consistent: Satisfies triangle inequality

- Integer: Easy to compare

- Fast: Simple arithmetic, no square roots

Euclidean Distance:

- Admissible: Never overestimates

- Less informed: Underestimates more than Manhattan

- Slower: Requires square root calculations

### Heuristic Comparison Example
Current State: Tile '8' at position (0,2)
Goal State:    Tile '8' at position (2,0)

Actual path needed:
(0,2)  (1,2)   (1,1)  (2,1)  (2,0) = 4 moves

Manhattan: |0-2| + |2-0| = 2 + 2 = 4  EXACT

Euclidean:  root(4 + 4) = root(8) = 2.83  UNDERESTIMATES

**Why Manhattan is better** :

- More accurate estimate of remaining moves

- Faster computation (no square roots)

- Integer values prevent precision issues

- Tighter bound -> fewer nodes expanded

**Assumptions Made**:
- Blank tile = 0 

- Goal state = (0,1,2,3,4,5,6,7,8) (ascending order)

- Move cost = 1 for all moves (uniform cost)

- Heuristics are admissible (never overestimate)

- State representation as tuple (hashable for set)

- Parent pointers stored for path reconstruction
